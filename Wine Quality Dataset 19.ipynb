```python
# calculatee skewness of "Volatile Acidity"
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew

# Example: Load your dataset
# df = pd.read_csv('your_dataset.csv')

# For demonstration, let's create a sample DataFrame similar to wine quality datasets
# Replace this with your actual data
df = pd.DataFrame({
    'volatile acidity': [0.34, 0.42, 0.36, 0.58, 0.51, 0.52, 0.30, 0.45, 0.49, 0.38]
})

# Calculate skewness
skewness_value = skew(df['volatile acidity'])

print(f"Skewness of volatile acidity: {skewness_value}")

# Plot distribution with seaborn
sns.histplot(df['volatile acidity'], kde=True)
plt.title(f"Distribution of Volatile Acidity (Skewness: {skewness_value:.2f})")
plt.xlabel('Volatile Acidity')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew

# Example: Load your dataset
# df = pd.read_csv('your_dataset.csv')

# For demonstration, let's create a sample DataFrame similar to wine quality datasets
# Replace this with your actual data
df = pd.DataFrame({
    'volatile acidity': [0.34, 0.42, 0.36, 0.58, 0.51, 0.52, 0.30, 0.45, 0.49, 0.38]
})

plt.figure(figsize=(11,6))
sns.histplot(data = df , x='volatile acidity', color='green', edgecolor='linen', bins=5, alpha=0.5, kde=True)
plt.xlabel('volatile acidity')
plt.ylabel('Density')
plt.vlines(df['volatile acidity'].mean(), ymin = 0, ymax = 4000, color='blue', linestyle='--', linewidth=2, label='Mean')
plt.vlines(df['volatile acidity'].median(), ymin = 0,ymax = 4000, color='red', linestyle='-.', linewidth=2, label='Median')
plt.legend()
plt.show()
```
